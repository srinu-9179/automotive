```python
# ==========================
# üß† Helmet Detection Setup
# ==========================

import os, requests

# STEP 1: Install dependencies
!pip install -q ultralytics opencv-python matplotlib

# STEP 2: Import required libraries
from ultralytics import YOLO
import cv2, matplotlib.pyplot as plt

print("‚úÖ Libraries installed and imported.")

# STEP 3: Try to download helmet detection model
url_helmet = "https://github.com/techzizou/Helmet-Detection-YOLOv8/releases/download/model/helmet_yolov8n.pt"
file_name = "helmet.pt"

print("\n‚¨áÔ∏è Downloading helmet detection model...")
r = requests.get(url_helmet, stream=True)

if r.status_code == 200:
    with open(file_name, "wb") as f:
        for chunk in r.iter_content(1024*64):
            f.write(chunk)
    print("‚úÖ Downloaded helmet.pt successfully.")
else:
    print(f"‚ö†Ô∏è Download failed (status {r.status_code}). Using backup model...")

    # STEP 4: Backup download (official YOLOv8n)
    backup_url = "https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8n.pt"
    r = requests.get(backup_url, stream=True)
    with open(file_name, "wb") as f:
        for chunk in r.iter_content(1024*64):
            f.write(chunk)
    print("‚úÖ Backup model downloaded (yolov8n.pt).")

# STEP 5: Check file size
size = os.path.getsize(file_name)
if size < 1000000:  # less than 1 MB = broken file
    raise ValueError("‚ùå Downloaded file is empty (0 bytes). Try rerunning the cell or check internet connection.")
else:
    print(f"üì¶ Model file size: {size/1e6:.2f} MB")

# STEP 6: Load the YOLO model
model = YOLO(file_name)
print("‚úÖ Model loaded successfully!")
print("üß© Model Classes:", model.names)
```

```python
# Step 1: Install dependencies
# Run this only once in terminal / Jupyter
# !pip install ultralytics opencv-python matplotlib

# Step 2: Import libraries
from ultralytics import YOLO
import cv2

# Step 3: Load the YOLOv8 helmet detection model
# (Download model beforehand or use yolov8n.pt as fallback)
model = YOLO("helmet.pt")  # or YOLO("yolov8n.pt")
print("‚úÖ Model loaded successfully!")

# Step 4: Initialize webcam
cap = cv2.VideoCapture(0)  # 0 = default camera

if not cap.isOpened():
    print("‚ùå Error: Cannot open webcam")
    exit()

print("üé• Press 'q' to quit the live detection window.")

# Step 5: Run detection loop
while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ö†Ô∏è Failed to grab frame")
        break

    # Run YOLO detection
    results = model(frame, conf=0.25)

    # Annotate the frame with detections
    annotated_frame = results[0].plot()

    # Display the output
    cv2.imshow("ü™ñ Helmet Detection (Press 'q' to quit)", annotated_frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Step 6: Release resources
cap.release()
cv2.destroyAllWindows()
print("‚úÖ Detection stopped and camera released.")
```
